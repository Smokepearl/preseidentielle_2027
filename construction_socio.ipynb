# Construction du dataset socio-démographique

Pour répondre au brief (« croiser les résultats avec des données sociales »), on rassemble
deux indicateurs INSEE par région :

- **Taux de chômage** (INSEE, 2022)
- **Niveau de vie médian** mensuel en euros (INSEE Filosofi 2021)

Les données INSEE sont publiées thème par thème (un fichier chômage, un fichier revenus...),
donc on les rassemble ici par région dans un seul tableau propre, avec les sources.

Sources :
- Chômage : INSEE, « Chômage dans les régions » (taux 2022)
- Niveau de vie : INSEE Filosofi 2021 (via observatoire des inégalités)

```python
import pandas as pd
import os
os.makedirs('data_clean', exist_ok=True)

# (code région, taux de chômage 2022 en %, niveau de vie médian mensuel 2021 en €)
donnees = {
    'Île-de-France':              (11, 6.9, 2101),
    'Centre-Val de Loire':        (24, 6.8, 1898),
    'Bourgogne-Franche-Comté':    (27, 6.3, 1896),
    'Normandie':                  (28, 6.9, 1873),
    'Hauts-de-France':            (32, 8.8, 1785),
    'Grand Est':                  (44, 7.1, 1913),
    'Pays de la Loire':           (52, 5.9, 1921),
    'Bretagne':                   (53, 5.8, 1937),
    'Nouvelle-Aquitaine':         (75, 6.5, 1893),
    'Occitanie':                  (76, 8.7, 1834),
    'Auvergne-Rhône-Alpes':       (84, 6.3, 1983),
    "Provence-Alpes-Côte d'Azur": (93, 8.2, 1902),
    'Corse':                      (94, 6.1, 1866),
}

df = pd.DataFrame(
    [(code, region, ch, nv) for region, (code, ch, nv) in donnees.items()],
    columns=['Code_region', 'Region', 'taux_chomage', 'niveau_vie_median'],
)
df.to_csv('data_clean/socio_demographie_regions.csv', sep=';', index=False, encoding='utf-8-sig')
print('Exporté : data_clean/socio_demographie_regions.csv')
df
```
